```python
train_df, test_df = train_test_split(df, test_size=0.2, random_state=42)
```

```python
train_df
#train_df.style.background_gradient(cmap='Blues')
```

```python
data = TfidfVectorizer(stop_words='english')
data_matrix = data.fit_transform(train_df['Text'])
```

```python
model = NearestNeighbors(n_neighbors=6, metric='cosine')
model.fit(data_matrix)
```

```python
def get_similar_products(product_id, N=5):
    index = df[df['ProductId'] == product_id].index[0]
    _, indices = model.kneighbors(data_matrix[index], n_neighbors=N+1)
    product_indices = indices[0][1:]
    return df.iloc[product_indices]
```
